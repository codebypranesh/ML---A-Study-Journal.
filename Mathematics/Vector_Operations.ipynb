# **Dot Product of 2 Vectors**

Returns a Scalar Value

```python
a = np.asarray([2,3])

b = np.asarray([4,4])

a_dot_b = np.dot(a,b)

print(a_dot_b)
```

```python
c = np.array([40, 20, 35])

d = np.array([53, 24, 68])

c_dot_d = np.dot(c, d)

print(c_dot_d)
```

Cross Product of 2 Vectors

```python
a = np.array([2, 3])

b = np.array([4, 4])

a_cross_b = np.cross(a, b)

print(a_cross_b)
```

```python
c = np.array([5, 10, 20])

d = np.array([18, 32, 50])

c_cross_d = np.cross(c, d)

print(c_cross_d)
```

Projection of "a" vector on "v" vector

projection of vector.png

```python
a = np.array([2, 5])
v = np.array([8, -6])

# magnitude squared of "v" (no need to take sqrt)
mag_squared = np.dot(v, v)

# projection of a onto v
proj_of_a_on_v = (np.dot(a, v) / mag_squared) * v

print('Projection of a vector on v vector =', proj_of_a_on_v)
```

```python
a = np.array([23, 45, 62])

v = np.array([45, 82, 67])

mag_squared = np.dot(v,v)

proj_of_a_on_v = (np.dot(a,v) /  mag_squared) * v

print('Projection of a vector on v vector =', proj_of_a_on_v)
```

This is known as the vector projection or more precisely:

Scalar projection times the unit direction of v (not necessarily unit vector if v isnt normalized)
