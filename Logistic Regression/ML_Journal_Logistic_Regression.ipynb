**Logistic Regression** : Cancer Prediction Dataset

About : Predict whether the cancer is benign or malignant.

Source : https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data/data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["Unnamed: 32","id"], axis=1, inplace=True)
data.head()
```

```python
data.diagnosis = [1 if value == "M" else 0 for value in data.diagnosis]
```

```python
data["diagnosis"] = data["diagnosis"].astype("category", copy = False)
data["diagnosis"].value_counts().plot(kind="bar")
```

```python
# Divide into target variable and predictors
y = data["diagnosis"] # our target variable
X = data.drop(["diagnosis"], axis=1) # our predictors
```

```python
# Normalization
```

```python
from sklearn.preprocessing import StandardScaler

#creating a scaler object
scaler = StandardScaler()

#fit the scaler into the data and transform the data
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Split the data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
# Train the model
```

```python
from sklearn.linear_model import LogisticRegression

# Creating the lr model
lr = LogisticRegression()

# Training the model
lr.fit(X_train, y_train)

# Predict the target variable on test data
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Evaluation of the model
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy : 2f}")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
